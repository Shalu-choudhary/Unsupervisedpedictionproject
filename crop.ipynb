#### KMeans algo used for unsupervised learning. in this outliers free algo are passed . the above algo used are not normalized 

<h4> To normalize the outliers

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

##### Object creation

```python
std=StandardScaler()   # change in standard form
# transformation complete
x_transformed=std.fit_transform(x)
```

```python
x_transformed    # standard form (SND distribution)
# standardScaler jisme change krega usme uska mean=0, standars devuation=1
```

```python
print("total no of clay sample:",x_transformed.shape[0])  
```

#### train testing spliting (only by X variable)

```python
train=x_transformed[:2100]
test=x_transformed[2100:0]
```

#### kmean--> do clustring/grouping .divided by similarity. by default it divide into 8 group (n_cluster=8)

```python
from sklearn.cluster import KMeans 
```

```python
obj=KMeans()   # object creation
# obj.fit---> only train the algo
# obj.fit_prediction----> do fit or predict both

# algo fitted on training data
obj.fit(train)
```

#### score find--->>

```python
clust=obj.labels_  # in output the number are between 1 to 7 anything
```

#### add new column which is predict by the kmeans algo

```python
testing_df=x[:2100]   # 2100 extract 
```

```python
testing_df['cluster']=clust
testing_df
```

```python
testing_df['Label']=df[:2100]['label']  # for label column
testing_df.head(40)

# N --- rainfall all are X 
```

```python
testing_df['cluster'].nunique()  # crate group 0 to 7 by kmeans algo
```

```python
testing_df['cluster'].value_counts()
```

```python
fifth_no_cluster=testing_df[testing_df['cluster']==5]
```

```python
print("Total point in 5th group:",fifth_no_cluster.shape[0])
```

```python
list(fifth_no_cluster['Label'].unique()) # these  are similar requirements crops are in one group
```

```python
obj.score(train,clust)   # pass two things
```

```python
from sklearn.metrics import silhouette_score # for calculate score in kmeans algo
```

```python
silhouette_score(train,clust)*100
# output range in silhouette score--> -1 to +1 
```
